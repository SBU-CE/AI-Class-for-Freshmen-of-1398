```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
```

```python
to_plot = df.corr()
```

```python
to_plot
```

```python
plt.figure(figsize=(9,9))
plt.imshow(to_plot.values, cmap='RdBu')
plt.xticks(range(len(to_plot)), to_plot.columns, rotation=45)
plt.yticks(range(len(to_plot)), to_plot.columns, rotation=45)
plt.clim(-1,1)
plt.colorbar()
plt.show()
```

```python
ages = df.Age[~df.Age.isna()].values
```

```python
ages
```

```python
survived = df['Survived'][~df.Age.isna()].values
```

```python
import numpy as np
bins = np.linspace(min(ages), max(ages), 9)
```

```python
plt.hist(ages[survived==1], density=True, alpha=0.4, bins=bins, label ='survived')
plt.hist(ages[survived==0], density=True, alpha=0.4, color='red', bins=bins, label='dead')

plt.legend()
```
